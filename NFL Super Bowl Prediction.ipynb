```python
import pandas as pd
import numpy as np
```

```python
Matches=pd.read_csv(r'/Users/macbook/Desktop/Super Bowl/Matches.csv')
Abb=pd.read_csv(r'/Users/macbook/Desktop/Super Bowl/Abbrevaitions.csv')
Abb
```

```python
Matches=Matches.merge(Abb, how="left" , left_on="Team1", right_on="Franchise")
Matches=Matches.merge(Abb, how="left" , left_on="Team2", right_on="Franchise")
Matches.head()
```

```python
Matches = Matches.rename(columns={'Commonly Used Abbreviations_x': 'Abb1'})
Matches = Matches.rename(columns={'Commonly Used Abbreviations_y': 'Abb2'})
Matches = Matches.drop('Franchise_x', axis=1)
Matches = Matches.drop('Franchise_y', axis=1)
Matches.head()
```

```python
sum((Matches["Abb2"]).isnull())
```

```python
AllTeam=pd.DataFrame()
Team_list=[]
for k in list((Matches["Abb1"]).unique()):
    for i in range(len(Matches)):
        if (Matches.Abb1.iloc[i]==k):
            Team_list.append(Matches.iloc[i])
    Team=pd.DataFrame(Team_list)
    Team["R_Team1_Score"]=Team["Team1_Score"].rolling(3).mean().shift(1)
    Team["R_Team1_Passing"]=Team["Team1_Passing"].rolling(3).mean().shift(1)
    Team["R_Team1_Rushing"]=Team["Team1_Rushing"].rolling(3).mean().shift(1)
    Team["R_Team1_Turnovers"]=Team["Team1_Turnovers"].rolling(3).mean().shift(1)

    AllTeam=pd.concat([AllTeam, Team] , axis=0, ignore_index=True)
    Team_list=[]
    
```

```python
AllTeam["R_Team2_Score"]=None
AllTeam["R_Team2_Passing"]=None
AllTeam["R_Team2_Rushing"]=None
AllTeam["R_Team2_Turnovers"]=None

for i in range(len(AllTeam)):
    case=AllTeam.Team2.iloc[i]
    w=AllTeam.Week.iloc[i]
    for j in range(len(AllTeam)):
        if (AllTeam.Team1.iloc[j]==case) and (AllTeam.Week.iloc[j]==w):
            print(j)
            AllTeam["R_Team2_Score"].iloc[i]=AllTeam["R_Team1_Score"].iloc[j]
            AllTeam["R_Team2_Passing"].iloc[i]=AllTeam["R_Team1_Passing"].iloc[j]
            AllTeam["R_Team2_Rushing"].iloc[i]=AllTeam["R_Team1_Rushing"].iloc[j]
            AllTeam["R_Team2_Turnovers"].iloc[i]=AllTeam["R_Team1_Turnovers"].iloc[j]
```

```python
elo["Week"]=None
for m in range(len(AllTeam)):
    for p in range(len(elo)):
        if ((AllTeam.Abb1.iloc[m]==elo.team1.iloc[p]) 
            and (AllTeam.Abb2.iloc[m]==elo.team2.iloc[p]) 
            and (AllTeam.Team1_Score.iloc[m]==elo.score1.iloc[p])
            and (AllTeam.Team2_Score.iloc[m]==elo.score2.iloc[p])):
            
            elo.Week.iloc[p]=AllTeam.Week.iloc[m]
        if ((AllTeam.Abb1.iloc[m]==elo.team2.iloc[p]) 
            and (AllTeam.Abb2.iloc[m]==elo.team1.iloc[p]) 
            and (AllTeam.Team1_Score.iloc[m]==elo.score2.iloc[p])
            and (AllTeam.Team2_Score.iloc[m]==elo.score1.iloc[p])):
            elo.Week.iloc[p]=AllTeam.Week.iloc[m]

                   
```
